```python
## THIS IS FOR PREPROCESSING RAW PDFS INTO pdfs that have specified pages and converts them into text files 
## this is used for each "ticker_number" can't be used on the entire folder at once
# Input(list of pages to pull [(startpage,endpage)], ticker of specific company (string), doc_title, page range to pull )
path = "/Users/Ethan Ohman/Desktop/Equity Research/"
new_addresses = "_data.pdf"
tickers = ["ADBE","AAPL","AMZN","BA","CMCSA","CSCO","CVX","DIS","FB","GOOGL","INTC","KO","LOW","MCD","MSFT","MU","NFLX","NKE",
           "PEP","SBUX","T","TGT","VZ","WMT","XOM"]

def get_pages(page_list,ticker, doc_title, amount):  #pages will be in a list(hopefully) - make a contingency for if they are not
    r_paper = pdf.PdfFileReader(open((path + ticker + "/" +  doc_title + ".pdf"),"rb"))
    splitter = pdf.PdfFileReader(open(path + "splitter_new.pdf","rb")).getPage(0) # research paper splitter "/-/-/-/-/-/-/-/-/-/-/"
    export_data = pdf.PdfFileWriter()  #now you have a writer for all said data
    for page_pairs in page_list:
        page_select = page_pairs[0]
        for x in range (amount):
            page_to_append = r_paper.getPage(page_select + (x-1))
            export_data.addPage(page_to_append)
            export_data.addPage(splitter)
        company = ""
        for tick in tickers: 
            try: 
                doc_title.index(tick) 
                company = ticker + "/" 
                break
            except:
                pass 
    export_data.write(open((path + company + ticker + "_data/" + doc_title + new_addresses ), "wb"))
pages_test= [(12,13),(113,114),(121,122),(154,155),(334,335),(514,515)]
ticker = "AAPL"
doc_title = "AAPL (1)"
get_pages(pages_test, ticker, doc_title, 1)


    
```

```python
def MS_parser(paper,type_doc): 
    if(type_doc == 'latest'):
        data_tuple = ms_latest(paper)
        return data_tuple
    if(type_doc == 'old'):
        data_tuple = ms_old(paper)
        return data_tuple
    if(type_doc == 'var1'):
        data_tuple = ms_var1(paper)
        return data_tuple

def ms_latest(paper):
    date = ''
    position = ''
    for index in range(0,2):
        try:
            check = paper[index].index(',')
            arr = paper[index].split(" ")
            date = "{0} {1} {2}".format(arr[0],arr[1],arr[2])
            break
        except: 
            pass 
    for index in range(8,13):
        try:
            check = paper[index].index("Stock Rating")
            position = paper[index + 2].strip()
            break
        except:
            pass
    data_tuple = ("Morgan Stanley", date, position)
    return data_tuple

def ms_old(paper):
    date=''
    position = ''
    for index in range(0,2):
        try:
            check = paper[index].index('|')
            date = paper[index][(check+1):-1]
            break
        except: 
            pass
    for index in range(96,106):
        if(paper[index] == "\n"):
            position = paper[index+1][:-1]
    print(date)
    print(position)
    data_tuple = ("Morgan Stanley", date, position)
    return data_tuple

def ms_var1(paper):
    date=''
    position=''
    for index in range(3,6):
        if(paper[index] == "\n"):
            date = paper[index+1][:-1]
            break
    for index in range(25,30):
        try:
            check = paper[index].index("Stock Rating")
            position = paper[index+1][:-1]
            break
        except: 
            pass
    data_tuple = ("Morgan Stanley", date, position)
    return data_tuple
```

```python
def Oppen_parser(paper):
    for d in range(2,4):
        try: 
            paper[d].index(',')
            date = paper[d][:-1]
            break
        except: 
            pass
    
    position = paper[10][:-1]
    if(position == ''):
        position = paper[11][:-1]
    
    data_tuple = ('Oppenheimer', date, position)
    return data_tuple
```

```python
def Evercore_parser(paper):
    for index in range(1,3):
        try:
            paper[index].index(',')
            date = paper[index]
            break
        except: 
            pass 
    for index in range(10,14):
        try:
            paper[index].index('TARGET PRICE')
            t_p = paper[index]
            break
        except:
            pass
    
    position = t_p[:(t_p.index('|'))].strip(' ')
    
    
    data_tuple = ('Evercore', date[:-1], position) 
    return data_tuple  
```

```python

def Jeff_parser(paper,type_doc):
    if(type_doc == 'latest'):
        data_tuple = jeff_latest(paper)
        return data_tuple
    if(type_doc == 'old'):
        data_tuple = jeff_older(paper)
        return data_tuple
    
    

def jeff_latest(paper):
    for index in range(5,8):
        try:
            paper[index].index(',')
            date = paper[index][:-1]
            break
        except:
            pass  
            
    position = paper[33][:-1]
    data_tuple = ('Jefferies',date,position)
    return data_tuple

def jeff_older(paper):
    for index in range(0,5):
        try:
            paper[index].index(',')
            date = paper[index][:-1]
            break
        except:
            pass
    position = paper[7][:-1]
#     for index in range(6,9):
#         if(paper[index][0] != ""):
#             print(position = paper[index])
#             position = paper[index]
#             break
    
    data_tuple = ('Jefferies',date,position)
    return data_tuple      
```

```python
def JP_parser(paper):
    for index in range(0,4):
        try:
            paper[index].index('0')
            date = paper[index][:-1]
            break
        except: 
            pass
    
    for index in range(len(paper)):
        try:
            paper[index].index('Price Target (')
            try:
                start_r = paper[index-2].index('$')
                end_r = paper[index-2].index('.')+3
                buffer = 0
            except: 
                start_r = paper[index-1].index('$')
                end_r = paper[index-1].index('.')+3
                buffer = 1         
            position = paper[index-4+buffer][:-1]
            break
        except: 
            pass
    
    data_tuple = ('JP Morgan',date,position)
    return data_tuple
```

```python
#make method in pandas which converts 


def evercoreParser(bankName,report):
    if(bankName!='EVERCORE ISI'):
        return ('DNE','DNE','DNE')
    #ADD INITIALIZERS
    position = 'DNE'
    price_target = 'DNE'
    #Position
    #Price Target
    for index in range(10):
        try:
            report[index].index('TARGET PRICE')
            info = report[index].split('|')
            position = info[0].strip()
            pt = info[1].split('$')
            pt = pt[1][:10]
            try:
                price_target = float(pt.strip())
            except:
                price_target = pt[1].strip()
            break
        except:
            pass       
    #Text
    text = ''
    for line in report:
        text += " " + line.strip()
    data_tuple = (position, price_target, text)
    return data_tuple

def oppenhiemerParser(bankName,report):
    
    if(bankName!='OPPENHIEMER AND CO'):
        return ('DNE','DNE','DNE')
    #ADD INITIALIZERS
    position = 'DNE'
    price_target = 'DNE'
    #Position
    for index in range(10):
        try:
            report[index].index('Stock Rating:')
            position = report[index+2].strip()
    #Price Target
            pt = report[index+3].split('12-18 mo. Price Target') 
            price_target = pt[1].strip()
            
            try:
                price_target = float(price_target)
                break
            except:
                break
        except:
            pass
    #Text
    text = ''
    for line in report:
        text += " " + line.strip()
        
    data_tuple = (position, price_target, text)
    return data_tuple
        

def jpmoragnParser(bankName,report):   #needs more work xd
    if(bankName!='JPMORGAN'):
        return ('DNE','DNE','DNE')
    #ADD INITIALIZERS
    position = 'DNE'
    price_target = 'DNE'
    #Position
    for index in range(10):
        try:
            report[index].index('Price (')
            position = report[index-2].strip()
        except:
            pass
        #Price Target
        try:
            report[index].index('Price Target')
            temp_list = report[index].split('$')
            price_target = float(temp_list[1].strip())
            break
        except:
            price_target = 0
    #Text
    text = ''
    for lines in report:
        text += " " + lines.strip()

    data_tuple = (position, price_target, text)
    return data_tuple

def jefferiesParser(bankName,report):      #for this one, there are two types of documents, the old version, and the new version
    if(bankName!='JEFFERIES'):
        return ('DNE','DNE','DNE')
    #ADD INITIALIZERS
    position = 'DNE'
    price_target = 'DNE'
    #//// NEW TYPE
    #Position
    oldreport = True
    for index in range(20):
        try:
            #Position
            report[index].index('RATING')
            temp_list = report[index].split('RATING')
            position = temp_list[1].strip()
            continue
        except:
            try:
                #Price Target
                report[index].index('PRICE TARGET (PT)')
                try:
                    temp_list = report[index].split('$')
                    price_target = temp_list[1].strip()
                    try:
                        price_target = float(price_target)
                    except:
                        pass
                    oldreport = False
                    break
                except:
                    temp_list = report[index-1].split('$')
                    price_target = temp_list[1].strip()
                    try:
                        price_target = float(price_target)
                    except:
                        pass
                    oldreport = False
                    break     
            except:
                pass
    if(oldreport):
        for index in range(10):
            try:
                #Position
                report[index].index(')')
                l = report[index].split(')')
                position = l[1].strip()
                continue
            except:
                try:
                    #Price Target
                    report[index].index('Price target')
                    p = report[index].split('$')
                    price_target = p[1].strip()
                    break
                except:
                    pass
        
    #Text 
    text = ''
    for lines in report:
        text += " " + lines.strip()   
    data_tuple = (position, price_target, text)
    return data_tuple

def morganStanParser(bankName,report):
    if(bankName!='MORGAN STANLEY'):
        return ('DNE','DNE','DNE')
    #ADD INITIALIZERS
    position = 'DNE'
    price_target = 'DNE'
    recent_report = False
    old_report = False
    try:
        report[0].index('|')
        old_report = True
    except:
        recent_report = True
    
    if(recent_report):
        for index in range(20):
            try:
                report[index].index('Stock Rating')
                try:
                    report[index+1].index('$')
                    temp_list = report[index+1].split('$')
                    position = temp_list[0][:15].strip()
                    price_target = temp_list[1][:7].strip()
                    try:
                        price_target = float(price_target)
                    except:
                        pass
                    break
                except:
                    temp_list = report[index+2].split('$')
                    position = temp_list[0][:15].strip()
                    price_target = temp_list[1][:7].strip()
                    try:
                        price_target = float(price_target)
                    except:
                        pass
                    break
            except:
                pass
    if(old_report):
        count = 0
        for index in range(20):
            try:
                report[index].index('Stock Rating')
                if(count != 1):
                    count =1
                    continue
                temp_list = report[index].split('Stock Rating')
                position = temp_list[1].strip()
                try:
                    report[index+2].index('$')
                    temp_list = report[index+2].split('$')
                    pt = temp_list[1].strip()
                    price_target = float(pt)
                    break
                except:
                    temp_list = report[index+4].split('$')
                    pt = temp_list[1].strip()
                    price_target = float(pt)
                    break
            except:
                pass
    
    #Text 
    text = ''
    for lines in report:
        text += " " + lines.strip()  
    data_tuple = (position, price_target, text)
    return data_tuple

    
def switch(argument,report):
    cases = {
        'JPMORGAN': jpmoragnParser(argument, report),
        'EVERCORE ISI': evercoreParser(argument, report),
        'MORGAN STANLEY': morganStanParser(argument, report),
        'OPPENHIEMER AND CO': oppenhiemerParser(argument, report),
        'JEFFERIES': jefferiesParser(argument, report)
    }
    data_tuple = cases.get(argument, ('DNE','DNE','DNE'))
    return data_tuple
    
def test_parsers(doc_name, bank_name):
    count = 0
    data_dictionary = {}
    data_dictionary['Position'] = []
    data_dictionary['Price Target'] = []
    data_dictionary['Text'] = []
    path = "/Users/Ethan Ohman/Desktop/Equity Research/AAPL/AAPL_data_txt/{0}.txt".format(doc_name)
    er_report = []
    with open(path,'r',errors='ignore') as file:
        total_collection = file.readlines()
        for line in total_collection:
            if(line == '\n'):
                continue
            else:
                try:
                    line.index("/-/-/-/-/-/-/-/-/-/-/NEWPAGE")
                    data_tuple = switch(bank_name, er_report)
                    # Position
                    data_dictionary['Position'].append(data_tuple[0])
                    #Text
                    data_dictionary['Text'].append(data_tuple[2])
                    # Price Target
                    data_dictionary['Price Target'].append(data_tuple[1])
                    er_report.clear()
                    
                except:
                    line.strip()
                    er_report.append(line[:-1])
                  
    print(data_dictionary)
    print(len(data_dictionary ['Price Target']))
    print(len(data_dictionary ['Position']))
    
# test_parsers('jpmorgan', 'JPMORGAN')    
# test_parsers('oppenhiemer', 'OPPENHIEMER AND CO')
# test_parsers('evercore', 'EVERCORE ISI')
# test_parsers('jefflatest', 'JEFFERIES')
# test_parsers('jeffold', 'JEFFERIES')
# test_parsers('jeffold2', 'JEFFERIES')
# test_parsers('jeffold3', 'JEFFERIES')
# test_parsers('morganrecent', 'MORGAN STANLEY')
# test_parsers('morganold', 'MORGAN STANLEY')
# test_parsers('morganrare', 'MORGAN STANLEY')
test_parsers('AAPL (9)_data', 'MORGAN STANLEY')




                    
        
        
```
